```python
group_by_columns=['año', 'departamento', 'grupo_de_cultivo']
aggregation_columns = {'count': ('grupo_de_cultivo', 'count'), 'avg_production': ('producci_n_t', 'mean')}
sort_by_column='avg_production'
grouped_table=eda.grouped_table(group_by_columns,aggregation_columns,sort_by_column)
grouped_table
```

- Los 'cultivos tropicales tradicionales' en el departamento del Valle del Cauca fueron los que mayor producción promedio tuvieron por año (2021).

```python
group_by_columns=['grupo_de_cultivo']
aggregation_columns = {'count': ('grupo_de_cultivo', 'count'), 'sum_production': ('producci_n_t', 'sum')}
sort_by_column='sum_production'
grouped_table=eda.grouped_table(group_by_columns,aggregation_columns,sort_by_column)
grouped_table
```

- Los grupos de cultivos tropicales tradicionales, frutales y raíces/tubérculos han registrado las mayores producciones en la historia, con 44,992,148 toneladas, 12,886,524 toneladas y 6,632,276 toneladas respectivamente.

```python
df = eda.data
df_filtered = df[df['año']==2022]
eda.update_dataframe(df_filtered)

group_by_columns=['año', 'grupo_de_cultivo']
aggregation_columns = {'count': ('grupo_de_cultivo', 'count'), 'avg_production': ('producci_n_t', 'mean')}
sort_by_column='avg_production'
grouped_table=eda.grouped_table(group_by_columns,aggregation_columns,sort_by_column)
grouped_table
```

- La producción promedio más alta se observa en los cultivos tropicales tradicionales con un promedio de 21,568.62 toneladas, seguidos de las oleaginosas con un promedio de 8,624.22 toneladas y las raíces y tubérculos con un promedio de 2,103.48 toneladas. Estos cultivos muestran una producción significativamente más alta en comparación con otros tipos de cultivos en el año 2022.
