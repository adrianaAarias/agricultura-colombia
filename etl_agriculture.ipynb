```python
import pandas as pd
from utils.func_utils import DataGovAPI, DataManager
import math
import os
```

# Data extraction

## Extracting data from 2007 to 2018 
link: https://dev.socrata.com/foundry/www.datos.gov.co/2pnw-mmge

```python
# Get dataset per batches
BATCH_SIZE = 10000
BASE_PATH = "data/raw/extraction_1"
END_SIZE = 200000
NUM_BATCH = int(math.ceil(END_SIZE/BATCH_SIZE))

api=DataGovAPI()
data_manager=DataManager()

for i_batch in range(NUM_BATCH):
    
    offset = BATCH_SIZE*i_batch + 1
    endpoint = f'https://www.datos.gov.co/resource/2pnw-mmge.json?$limit={BATCH_SIZE}&$offset={offset}'
    api.load_data(endpoint)
    df=api.get_dataframe()
    data_manager.save_to_csv(df, f'{BASE_PATH}/agriculture_col_{i_batch}.csv')
```

## Extracting data from 2019 to 2022
link: https://dev.socrata.com/foundry/www.datos.gov.co/uejq-wxrr

```python
# Get dataset per batches
BATCH_SIZE = 10000
BASE_PATH = "data/raw/extraction_2"
END_SIZE = 92000
NUM_BATCH = int(math.ceil(END_SIZE/BATCH_SIZE))

api=DataGovAPI()
data_manager=DataManager()

for i_batch in range(NUM_BATCH):
    
    offset = BATCH_SIZE*i_batch + 1
    endpoint = f'https://www.datos.gov.co/resource/uejq-wxrr.json?$limit={BATCH_SIZE}&$offset={offset}'
    api.load_data(endpoint)
    df=api.get_dataframe()
    data_manager.save_to_csv(df, f'{BASE_PATH}/agriculture_col_{i_batch}.csv')
```

# Data transformation

```python
# Define the mapping of columns
column_mapping = {
    'c_digo_dane_departamento': 'c_d_dep',
    'departamento': 'departamento',
    'c_digo_dane_municipio': 'c_d_mun',
    'municipio': 'municipio',
    'grupo_cultivo': 'grupo_de_cultivo',
    'subgrupo': 'subgrupo_de_cultivo',
    'cultivo': 'cultivo',
    'desagregaci_n_cultivo': 'desagregaci_n_regional_y',
    'a_o': 'a_o',
    'periodo': 'periodo',
    'rea_sembrada': 'rea_sembrada_ha',
    'rea_cosechada': 'rea_cosechada_ha',
    'producci_n': 'producci_n_t',
    'rendimiento': 'rendimiento_t_ha',
    'ciclo_del_cultivo': 'ciclo_de_cultivo',
    'estado_f_sico_del_cultivo': 'estado_fisico_produccion',
    'c_digo_del_cultivo': 'c_digo_del_cultivo',
    'nombre_cient_fico_del_cultivo': 'nombre_cientifico'
}

# Paths to the folders containing the CSV files
folders = ['data/raw/extraction_1', 'data/raw/extraction_2']

# Read and map CSV files from both folders, then concatenate them
all_dfs = []
for folder in folders:
    list_files = os.listdir(folder)
    for file in list_files:
        if file.endswith('.csv'):
            all_dfs.append(pd.read_csv(os.path.join(folder, file)).rename(columns=column_mapping))


# Concatenate all DataFrames into a single DataFrame
final_df = pd.concat(all_dfs, ignore_index=True)

print("All CSV files have been concatenated")
```

```python
data_manager=DataManager()
data_manager.fix_string_case(final_df)
if 'departamento' in final_df.columns:
            final_df['departamento'] = final_df['departamento'].replace({
                'San andres y providencia': 'San andres y providencias',
                'Archipielago de san andres, providencia y santa catalina': 'San andres y providencias',
                 'Narino':'Nariño'     
            })
final_df=final_df.rename(columns={'a_o': 'año'})
```

```python
data_manager.save_to_csv(final_df, 'data/transformed/agriculture.csv')
```
